-Create a graph that shows the null and hypotheses, the type 1 and 2 errors that looks like this:



```
import numpy as np
from scipy.ndimage import label
from scipy.stats import norm, alpha
from matplotlib import pyplot as plt


mu, sigma = 3, 2
s = np.random.normal(mu, sigma, 10000)
#just want to see my sample
#print(s)
#bins to incraese the efficiency of my histogram and density=true to normalize the histogram to fit under the curve
#refernce for this https://www.youtube.com/watch?v=2nMQnjQvMv0
plt.hist(s,bins=50, density=True, zorder=10, label="NULL Hypothesis Hist")


#defining the x axis limits
(xmin, xmax) = plt.xlim()

#to have evenly spaced numbers
x_limts = np.linspace(xmin, xmax)
#to plot the curve
probab_dist = norm.pdf(x_limts, mu, sigma)
plt.plot(x_limts, probab_dist,   color='blue', zorder=15, label="NULL Hypothesis PDF")

#to plot the sample normal distribution
sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean, sample_sigma, 200)
count, bins, ignored = plt.hist(sample, 30, alpha=0.2, density=True, zorder=5, label="Alternate Hypothesis Hist")
# i am taking the variable new_probab_dist
new_probab_dist = norm.pdf(x_limts, sample_mean, sample_sigma)
plt.plot(x_limts, new_probab_dist,color='red', zorder=15, label="Alternate Hypothesis PDF")

#The null hypothesis(the first distribution that had to be plotted) is a normal distribution with two tailed errors here I am assuming a significance level of 0.05
significance_level_pop=norm.interval(0.95,loc=mu,scale=sigma)
print('the sample confidence interval values for the population density function are',(significance_level_pop))
plt.axvline(significance_level_pop[0],color='purple', label="Significance Level 0.05",zorder=28)
plt.axvline(significance_level_pop[1],color='purple',zorder=28)
val1 =np.linspace(significance_level_pop[0],significance_level_pop[1])

#to take the 95% confidence interval for the sample density
conf_int_sam=norm.interval(0.95,loc=sample_mean,scale=sample_sigma)
print('the sample confidence interval values are sample density function',(conf_int_sam))
plt.axvline(conf_int_sam[0],color='g', label="Confidence Interval",zorder=28)
plt.axvline(conf_int_sam[1],color='g',zorder=28)
plt.show()
```

-Compute the z-score, which is equal to (sample mean – mu)/sigma. What is the p-value and do you reject it?



```
import numpy as np
from scipy.ndimage import label
from scipy.stats import norm, alpha
from matplotlib import pyplot as plt


mu, sigma = 3, 2
s = np.random.normal(mu, sigma, 10000)
#just want to see my sample
#print(s)
#bins to incraese the efficiency of my histogram and density=true to normalize the histogram to fit under the curve
#refernce for this https://www.youtube.com/watch?v=2nMQnjQvMv0
plt.figure(figsize=(10,6))
plt.hist(s,bins=50, density=True, zorder=10, label="NULL Hypothesis Hist")


#defining the x axis limits
(xmin, xmax) = plt.xlim()

#to have evenly spaced numbers
x_limts = np.linspace(xmin, xmax)
#to plot the curve
probab_dist = norm.pdf(x_limts, mu, sigma)
plt.plot(x_limts, probab_dist,   color='blue', zorder=15, label="NULL Hypothesis PDF")

#to plot the sample normal distribution
sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean, sample_sigma, 200)
count, bins, ignored = plt.hist(sample, 30, alpha=0.2, density=True, zorder=5, label="Alternate Hypothesis Hist")
# i am taking the variable new_probab_dist
new_probab_dist = norm.pdf(x_limts, sample_mean, sample_sigma)
plt.plot(x_limts, new_probab_dist,color='red', zorder=15, label="Alternate Hypothesis PDF")

#The null hypothesis(the first distribution that had to be plotted) is a normal distribution with two tailed errors here I am assuming a significance level of 0.05
significance_level_pop=norm.interval(0.95,loc=mu,scale=sigma)
print('the sample confidence interval values for the population density function are',(significance_level_pop))
plt.axvline(significance_level_pop[0],color='purple', label="Significance Level 0.05",zorder=28)
plt.axvline(significance_level_pop[1],color='purple',zorder=28)
val1 =np.linspace(significance_level_pop[0],significance_level_pop[1])

#to take the 95% confidence interval for the sample density
conf_int_sam=norm.interval(0.95,loc=sample_mean,scale=sample_sigma)
print('the sample confidence interval values are sample density function',(conf_int_sam))
plt.axvline(conf_int_sam[0],color='g', label="Confidence Interval",zorder=28)
plt.axvline(conf_int_sam[1],color='g',zorder=28)

#just to colour the interval for type 1 and Type 2 error

# type 2 error happens when the null hypothesis is incorrect but the findings are not enough to reject it and hence  the null Hypothesis bis accepted
plt.fill_between(val1,0,norm.pdf(val1, sample_mean, sample_sigma), color='yellow',alpha=0.4, zorder=20, label="Type 2 Error")
val2 = np.linspace(xmin, significance_level_pop[0])
val3 = np.linspace(significance_level_pop[1], xmax)
# type 1 error happens when the null hypothesis is correct but we do not have enough evidende to support it and hence we reject the null Hypothesis
plt.fill_between(val2,0,norm.pdf(val2, mu, sigma), color='black',alpha=0.8, zorder=26, label="Type 1 Error")
plt.fill_between(val3,0,norm.pdf(val3, mu, sigma), color='black',alpha=0.8, zorder=26)





# to calculte the z score

score_z=(sample_mean-mu)/sigma
print('the z-score is',(score_z))

# to calculte the p-value (Since we are taking a two tailed Hypothesis (reason: as metioned in the plot below the null hypothesis show 2 tailed)
p_value=norm.sf(abs(score_z)*2)
print(p_value)

#decision to reject or accept
print('Since the p value of {} is greater that 0.05 we cannot reject the Hypothesis'.format(p_value))

plt.legend()
plt.title("Hypothesis Testing Anusha")
plt.ylabel("Density")
plt.show()
```
